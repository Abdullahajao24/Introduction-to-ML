```python
x = heart.drop("DEATH_EVENT", axis=1)
y = heart['DEATH_EVENT']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, f1_score, confusion_matrix, classification_report

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=25)
```

```python
def scores(y_test, pred):
    print(f"Accuracy:: {accuracy_score(y_test, pred)}")
    print(f"Precision:: {precision_score(y_test, pred)}")
    print(f"F1 score:: {f1_score(y_test, pred)}")
    print(f"Confusion Matrix:: {confusion_matrix(y_test, pred)}")
    print(f"Classification Report:: {classification_report(y_test, pred)}")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
pred = rf.predict(x_test)

scores(y_test, pred)
```

```python
feature_importance = rf.feature_importances_

features = x.columns
df = pd.DataFrame({
    'Features': features,
    'Importance': feature_importance
})

df = df.sort_values('Importance', ascending=True)

plt.figure(figsize=(10, 8))
plt.bar(df['Features'], df['Importance'])
plt.title("Features Importance")
plt.xticks(rotation=45)
plt.xlabel("Features")
plt.ylabel("Importance")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

dc = DecisionTreeClassifier()
dc.fit(x_train, y_train)
pred = dc.predict(x_test)

scores(y_test, pred)
```

```python
feature_importance = dc.feature_importances_

features = x.columns
df = pd.DataFrame({
    'Features': features,
    'Importance': feature_importance
})

df = df.sort_values('Importance', ascending=True)

plt.figure(figsize=(10, 8))
plt.bar(df['Features'], df['Importance'])
plt.title("Features Importance")
plt.xticks(rotation=45)
plt.xlabel("Features")
plt.ylabel("Importance")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
pred = lr.predict(x_test)

scores(y_test, pred)
```

```python
from sklearn.model_selection import cross_val_score

# Using the same RandomForestClassifier and data from above

# Perform cross-validation
cv_scores = cross_val_score(rf, x, y, cv=10)

# Print the cross-validation scores
print(f"CV scores: {cv_scores}")

# Check if there might be overfitting
trainn = rf.predict(x_train)
train_accuracy = accuracy_score(y_train, trainn)
if train_accuracy > cv_scores.mean():
    overfitting_difference = train_accuracy - cv_scores.mean()
    if overfitting_difference > 0.1:  # arbitrary threshold
        print(f"Model may be overfitting. Performance difference: {overfitting_difference}")
```

```python
from sklearn.model_selection import KFold, GridSearchCV
```

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
```

```python
param_grid = {
    'n_estimators': [50, 100, 150, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
rff = RandomForestClassifier()
grid_search = GridSearchCV(estimator=rff, param_grid=param_grid, cv=kfold, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(x_train, y_train)
```

```python
best_params = grid_search.best_params_
print(best_params)
```

```python
best_rf = RandomForestClassifier(**best_params)
best_rf.fit(x_train, y_train)
best_rf_pred = best_rf.predict(x_test)
```

```python
scores(y_test, best_rf_pred)
```
